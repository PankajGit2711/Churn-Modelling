```
data = pd.read_csv('Churn_Modelling.csv')
x = data.iloc[:, 3:13].values
y = data.iloc[:, 13].values
```

```
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label1 = LabelEncoder()
x[:, 1] = label1.fit_transform(x[:, 1])
label2 = LabelEncoder()
x[:, 2] = label2.fit_transform(x[:, 2])
ohe = OneHotEncoder(categorical_features = [1])
x = ohe.fit_transform(x).toarray()
x = x[:, 1:]
```

```
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2)
```

```
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
xtrain = scale.fit_transform(xtrain)
xtest = scale.transform(xtest)
```

```
ann = Sequential()
```

```
#Input Layer & First Hidden Layer
ann.add(layer = Dense(output_dim = 6, init = 'uniform', activation = 'relu', input_dim = 11))
#Second Hidden Layer
ann.add(layer = Dense(output_dim = 6, init = 'uniform', activation = 'relu'))
#Output Layer
ann.add(layer = Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))
```

```
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```
ann.fit(xtrain, ytrain, batch_size = 10, epochs = 100)
```

```
predict = ann.predict(xtest)
predict = (predict > 0.5)
```

```
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(ytest, predict))
print('\n')
print("Accuracy Of ANN is: ", accuracy_score(ytest, predict))
```
